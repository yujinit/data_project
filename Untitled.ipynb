### 데이터 전처리

```python
obj_col = train.dtypes[train.dtypes == "object"].index
```

```python
for i in obj_col :
    train.loc[train[i].isnull(),i] = "None"
```

```python
num_col = []
for col,dtype in train.dtypes.items():
    if dtype == "int" or dtype == "float" :
        if col != "Age":
            num_col.append(col)
num_col            
```

```python
for j in num_col :
    train.loc[:,j]=train[j].fillna(0)
```

```python
import math
age_mean = math.trunc(train["Age"].mean())

train.loc[:,"Age"]=train["Age"].fillna(age_mean)
```

```python
train.isnull().sum().sort_values()
```

```python
train_dummies = pd.get_dummies(train)
train_dummies.head()
```

```python
plt.figure(figsize=(15,8))
train_dummies.corr()["ProdTaken"].sort_values(ascending=False).plot(kind="bar")
```

여권 여부가 ProdTaken(여행 패키지 신청)과 상관계수가 큰 변수<br>Age 가 클수록 ProdTaken 은 감소

```python
train["ProdTaken"].value_counts(normalize=True).rename("proportion").reset_index().rename(columns={"index":"ProdTaken"})
```

```python
train.groupby("Passport")["ProdTaken"].value_counts(normalize=True).rename('proportion').reset_index()
```

```python
sns.barplot(x="Passport",y="proportion",hue="ProdTaken",data=train.groupby("Passport")["ProdTaken"].value_counts(normalize=True).rename('proportion').reset_index())
```
